```python
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']
# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
for col in categorical_features:
    print(df[col].value_counts(normalize=True) * 100)
    print('---------------------------')
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(df.corr(),annot= True,linewidths=1, linecolor="white", cbar=True,
cmap = "CMRmap",xticklabels="auto", yticklabels="auto")
```

```python
ax=df['Restaurant Name'].value_counts().head(15).plot.bar(figsize =(12,6))
ax.set_title("Top 15 Restarurents with maximum outlets")
for i in ax.patches:
    ax.annotate(i.get_height(), (i.get_x() * 1.005, i.get_height() * 1.005))
```

```python
max_votes =df.Votes.sort_values(ascending=False).head(10)
df.loc[df['Votes'].isin(max_votes)][['Restaurant Name','Votes']]
df.loc[df['Votes'].isin(max_votes)][['Restaurant Name','Votes']].plot.bar(x='Restaurant Name', y='Votes',figsize = (10,6), color='green')
```

```python
df.plot.scatter(x='Average Cost for two',y='Aggregate rating',figsize=(10,6), color='blue', title="Cost vs Agg Rating")
```

```python
Delhi = df[(df.City == 'New Delhi')]
plt.figure(figsize=(12,6))
sns.barplot(x=Delhi.Locality.value_counts().head(10), y=Delhi.Locality.value_counts().head(10).index)

plt.ylabel(None);
plt.xlabel('Number of Resturants')
plt.title('Resturants Listing on Zomato');
```

```python
ConnaughtPlace = Delhi[(Delhi.Locality.isin(['Connaught Place'])) & (Delhi['Rating text'].isin(['Excellent','Very Good']))]

ConnaughtPlace = ConnaughtPlace.Cuisines.value_counts().head(5).reset_index()

## Extracing all the cuisens in a single list
cuisine = []
for x in ConnaughtPlace['index']: 
  cuisine.append(x)

cuisine
```

```python
max_delivery = Delhi.Locality.value_counts().head(10)
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12,6))
ax = sns.countplot(y= "Locality", hue="Has Online delivery", data=Delhi[Delhi.Locality.isin(max_delivery.index)])
plt.title('Resturants Online Delivery');
```
