```python
print("The percentage of null values in the columns consisting of null values:\n")
for i in ball_data.columns:
    null_rate = ball_data[i].isna().sum() / len(ball_data) * 100 
    if null_rate > 0 :
        print("null rate of {} column: {}%".format(i,round(null_rate,3)))
```

```python
ball_data['Avg'] = ball_data['total_runs'] / (ball_data['inning'])
ball_data['SR'] = ball_data['total_runs'] / (ball_data['ball'])
```

```python
numeric_features = [feature for feature in ball_data.columns if ball_data[feature].dtype != 'O']
categorical_features = [feature for feature in ball_data.columns if ball_data[feature].dtype == 'O']
# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(ball_data.corr(),annot= True,linewidths=1, linecolor="white", cbar=True,
cmap = "CMRmap",xticklabels="auto", yticklabels="auto")
```

```python
ax=ball_data['batsman'].value_counts().head(15).plot.bar(figsize =(12,6))
ax.set_title("Top 15 Batsman with most runs")
for i in ax.patches:
    ax.annotate(i.get_height(), (i.get_x() * 1.005, i.get_height() * 1.005))
```
